```python
so2 = df.groupby(["State"])["SO2"].max()
```

```python
so2.plot.bar(subplots = True, figsize=(10,8))
```

```python
no2 = df.groupby(["State"])["NO2"].max()
```

```python
no2.plot.pie(subplots = True, figsize=(10,8))
```

```python
y = df.groupby(["State"])["PM10"].max()
```

```python
y.plot.bar(subplots = True, figsize=(10,8))
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.head(200000).groupby(df["Date"].dt.floor('d'))[[ "PM10", "NO2","SO2"]].max().plot()
```
